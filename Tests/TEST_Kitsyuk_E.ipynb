Есть два файла, src_survey и src_OnOff. 
В таблице src_survey результат опроса респондентов - участников исследования, целью которого является проверка работы оборудования и регистрации респондентов на пиплметре (измерительное устройство для фиксации телесмотрения).

Далее, результат опроса сравнивается с данными регистрации респондента на пиплметре во время опроса, результат сравнения фиксируется в таблице (см. описание полей)

1.	Количество ситуаций (абс, %), когда прибор фактически выключен (данные в таблице src_OnOff), а респондент утверждает, что включен


2.	Количество ситуаций (абс, %), когда прибор фактически включен (данные в таблице src_OnOff), а респондент утверждает, что выключен


3.	Выделить ДХ с размером более 1 человека, среди них ситуации, телевизор фактически включен, респондент по данным опроса находится в комнате с включенным телевизором, регистрации нет, а у других респондентов в этом ДХ регистрация на этом телевизоре есть. Необходимо сгруппировать результат в зависимости от размера ДХ. \

Описание полей:
●	статус телевизора по утверждению респондента: tv state

●	размер ДХ: Кол-во человек в семье 4+

●	наличие фактической регистрации на пиплметре: see

●	наличие респондента в комнате с телевизором согласно опросу: display

●	идентификатор респондента: button

## Исследовательский анализ

### Количество ситуаций (абс, %), когда прибор фактически выключен (данные в таблице src_OnOff), а респондент утверждает, что включен

```python
# Выдел
device_off = survey_real[((survey_real['device_state'] == 'OFF') & (survey_real['tv_state']  == 'включён'))\
                & ((survey_real['survey_start'] > survey_real['begin']) & (survey_real['survey_start'] < survey_real['end']))]\
                .reset_index(drop=True)
device_off
```

```python
print('Общее количество ситуаций:', len(survey_real))
print('Количество ситуаций, когда прибор фактически выключен, а пользователь утверждает, что включен:', len(device_off))    
print(f'(% таких ситуаций от общего количества: {round(len(device_off) * 100 /len(survey_real), 2)}%)')
```

**Вывод:** Найдена всего лишь одна ситуация, когда респондент утверждает, что телевизор включен, а он фактически выключен. Можно заметить, что через 13 секунд после начала опроса телевизор включили.
Таким образом, на основании опросов респондентов, можем считать, что при фактически выключенном телевизоре пиплметр работает корректно.

###	Количество ситуаций (абс, %), когда прибор фактически включен (данные в таблице src_OnOff), а респондент утверждает, что выключен

```python
device_on = survey_real[((survey_real['device_state'] == 'ON') & (survey_real['tv_state']  == 'выключен'))\
                & ((survey_real['survey_start'] > survey_real['begin']) & (survey_real['survey_start'] < survey_real['end']))]\
                .reset_index()
device_on
```

```python
print('Количество ситуаций, когда прибор фактически включен, а пользователь утверждает, что выключен:', len(device_on))    
print(f'(% таких ситуаций от общего количества: {round(len(device_on) * 100 /len(survey_real), 2)}%)')
```

**Вывод:** Найдено всего лишь 3 ситуации, когда респондент утверждает, что прибор выключен, а он фактически включён. 
Стоит отметить, что 2 случая можно расценивать как один, так как они произошли в рамках опроса одного домохозяйства.

Таким образом, количество несоответствий ничтожно мало, и мы можем считать, что при фактически включённом телевизоре пиплметр тоже работает корректно.

###	Выделить ДХ с размером более 1 человека, среди них ситуации, телевизор фактически включен, респондент по данным опроса находится в комнате с включенным телевизором, регистрации нет, а у других респондентов в этом ДХ регистрация на этом телевизоре есть. Необходимо сгруппировать результат в зависимости от размера ДХ. 

```python
# Выделим ДХ с размером более 1 человека
data = survey_real[survey_real['people_cnt'] >1].reset_index(drop=True)
data
```

```python
# Далее выделим ситуации, когда телевизор включен, респондент по данным опроса в комнате с включенным телевизором,
# регистрации нет. Так мы выделим ДХ, которые потенциально могут подойти.
data_2 = data[(data['device_state'] == 'ON') & (data['display'] == 'в комнате')\
                & (data['tv_state'] == 'включён') & (data['see'] == 'нет регистрации')].reset_index(drop=True)
data_2
```

```python
# Выведем список номеров ДХ, которые потенциально могут подойти.
data_x = data_2['passport_number'].unique()
data_x
```

```python
# Рассмотрим все ситуации в вышеуказанных ДХ
data_3 = data[(data['passport_number'] == 1205697) | (data['passport_number'] == 1208135)\
            | (data['passport_number'] == 1210425) | (data['passport_number'] == 1210930)].reset_index(drop=True)
data_3
```

```python
# Выделим ДХ, в которых одновременно есть респонденты и с регистрацией, и без. 
# Т.е. сгруппировав данные по ДХ и телевизору, смотрим когда уникальных значений в столбце see больше, чем одно.
data_4 = data_3.groupby(['passport_number', 'unit_number']).agg({'see':'nunique'})
data_4 = data_4[data_4['see'] !=1]
data_4
```

```python
# Выведем заданные в задаче ситуации: телевизор фактически включён, респондент по данным опроса в комнате
# с включенным телевизором, регистрации нет. При этом только в двух вышеуказанных ДХ одновременно есть другие респонденты
# с регистрацией на этом телевизоре. 
final_situations = data_3[((data_3['passport_number'] == 1210425) | (data_3['passport_number'] == 1210930))\
            & ((data_3['device_state'] == 'ON') & (data_3['display'] == 'в комнате') & (data_3['tv_state'] == 'включён')\
            & (data_3['see'] == 'нет регистрации'))].reset_index(drop=True)
final_situations
```

```python
# Сгруппируем количество ситуаций по размеру ДХ
final_passport_numbers = final_situations.groupby(['passport_number', 'people_cnt']).agg({'code':'count'}).reset_index()
final_passport_numbers.columns = ['passport_number', 'people_cnt', 'situations_cnt']
final_passport_numbers
```

**Вывод:** Выделено два ДХ (1210425 и 1210930), в которых имели место ситуации, когда телевизор фактически включён, респондент по данным опроса в комнате с включенным телевизором, регистрации нет, а у других респондентов регистрация на этом телевизоре есть. Количество таких ситуаций: для 1210425 - 43 ситуаций, для 1210930 - 18 ситуаций.
